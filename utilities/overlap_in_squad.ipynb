```python
#Reference paper: Learning to Ask: Neural Question Generation for Reading Comprehension

# This notebook is to validate and correct pre-processing for the below statement from the paper :
# about 6.67% of the pairs have no non- stop-words in common, and this is mostly because of the answer offset error introduced during an- notation. 
# Therefore, we prune the training set based on the constraint: the sentence-question pair must have at least one non-stop-word in common.
```

```python
import linecache
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('punkt')
```

```python
stopWords = set(stopwords.words('english'))
```

```python
!cd /home/sdatta/git/drew/model
```

# Train dataset

```python
list_valid_records=[]
with open('data/src-train.txt', 'rb') as f:
    cntr=0
    for line in f:
        tgt_line=str(linecache.getline('data/tgt-train.txt', cntr+1))
        tgt_list=word_tokenize(tgt_line)
        src_line=str(linecache.getline('data/src-train.txt', cntr+1))
        src_list=word_tokenize(src_line)
        cnt_words_overlap=len(set(set(tgt_list) & set(src_list) )  - stopWords)
        if cnt_words_overlap >0:
            list_valid_records.append(cntr)
        else:
            print("======cntr:",cntr)
            print("cnt_words_overlap",cnt_words_overlap)
            print("Src ",src_line)
            print("Tgt ",tgt_line)
        cntr=cntr+1
```

```python
src_filtered = open('data/src-train-filtered.txt','w') 
tgt_filtered = open('data/tgt-train-filtered.txt','w') 

for rec_num in list_valid_records:
    src_line=linecache.getline('data/src-train.txt', rec_num+1)
    tgt_line=linecache.getline('data/tgt-train.txt', rec_num+1)
    src_filtered.write(src_line)
    tgt_filtered.write(tgt_line)
src_filtered.close()
tgt_filtered.close()
    
```
